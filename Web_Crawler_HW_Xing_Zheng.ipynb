```python
import requests

req = requests.get('https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population')
```

```python
webpage = req.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(webpage, 'html.parser')
```

```python
# print(soup.prettify())
```

```python
paragraph = soup.find('p', attrs={"class":False})
```

```python
paragraph.find_all('a', attrs={"title":True})
```

```python
data = {"title":[], "href":[]}
for link in paragraph.find_all('a', attrs={"title":True}):
    data["title"].append(link["title"])
    data["href"].append(link["href"])
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df
```

```python
webpages = []
head = "https://en.wikipedia.org"
for href in data["href"]:
    link = head + href
    req = requests.get(link)
    webpage = req.text
    webpages.append(webpage)
```

```python
req = requests.get('https://en.wikipedia.org/wiki/List_of_wine-producing_regions#Countries')
webpage = req.text
soup = BeautifulSoup(webpage, 'html.parser')
# paragraph = soup.find('a', attrs={"class":False})
# paragraph.find_all('link', attrs={"href":True})
link = soup.find_all('link', attrs = {"href": True})

data = {"rel":[], "href":[]}
for links in soup.find_all('link', attrs = {"href": True}):
    data["rel"].append(links["rel"])
    data["href"].append(links["href"])
df = pd.DataFrame(data)    
```
